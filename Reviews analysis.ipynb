```python
import pandas as pd
import string

# Step 2: Define a function to clean text (convert to lowercase and remove punctuations)
def clean_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    return text

# Step 3: Apply the function to the specific column (replace 'your_column' with your actual column name)
df['Review_clean'] = df['Review'].apply(lambda x: clean_text(x) if isinstance(x, str) else x)

# Step 4: Save the cleaned data to a new Excel file
df.to_excel('cleaned_file.xlsx', index=False)

print("Text cleaning completed and saved to 'cleaned_file.xlsx'.")
```

```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Initialize VADER sentiment intensity analyzer
analyzer = SentimentIntensityAnalyzer()

# Function to determine sentiment using VADER
def get_vader_sentiment(review):
    sentiment_score = analyzer.polarity_scores(review)
    # Classify sentiment based on compound score
    if sentiment_score['compound'] >= 0.05:
        return 'Positive'
    elif sentiment_score['compound'] <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

# Step 2: Apply VADER sentiment analysis to the cleaned text
df['Sentiment'] = df['Review_clean'].apply(get_vader_sentiment)

# Step 3: Save the updated DataFrame back to Excel
df.to_excel("hotel_reviews_with_sentiment_vader.xlsx", index=False)

print("Sentiment analysis complete using VADER! Results saved to 'hotel_reviews_with_sentiment_vader.xlsx'.")
```
